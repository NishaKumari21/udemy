```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

kmeans.inertia_:
<br>
This is a built-in attribute of the KMeans object in scikit-learn.
It returns the WCSS, i.e., the sum of squared distances from each point to its assigned cluster centroid.
Lower WCSS means tighter and more compact clusters.

```python
wcss
```

```python
# Step 4: Plot the Elbow Graph
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method to find Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
# Step 5: Apply KMeans with optimal K (e.g., K=4)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(x)
```

```python

centers=kmeans.cluster_centers_
centers
# Gives the coordinates of the centroids (center points of each cluster).
```

```python

labels=kmeans.labels_
labels
# Tells which cluster each data point belongs to
```

```python
plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis',marker='o',edgecolors='k')
plt.scatter(centers[:,0],centers[:,1],c='red',marker='*',edgecolors='k',label='centroids')
plt.title('Kmeans')
plt.xlabel('feature1')
plt.ylabel('feature1')
plt.show()
```

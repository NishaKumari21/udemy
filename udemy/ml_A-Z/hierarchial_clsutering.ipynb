ðŸ”¹ method='ward' â€“ Linkage Criterion
ward method ka kaam hai ki:
Jab clusters ban rahe hoon, toh jitni ho sake kam internal difference (variance) ho unmein.
Yeh method compact aur circular shape wale clusters ke liye best hai.
